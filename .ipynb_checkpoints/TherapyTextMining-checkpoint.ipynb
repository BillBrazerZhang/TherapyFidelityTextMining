```python
import numpy
import urllib.request
import scipy.optimize
import random
from collections import defaultdict
import nltk
import string
from nltk.stem.porter import *
from sklearn import linear_model

#nltk.download("stopwords")
from nltk.corpus import stopwords

def parseData(fname):
    with open(fname, encoding='utf-8', errors='ignore') as f:
        data = f.read()
    tokenizer = nltk.tokenize.TreebankWordTokenizer()
    data = tokenizer.tokenize(data)
    return data

stopWords = set(stopwords.words("english"))


print("Reading data...")
data = parseData('T1.txt')
print(data)
print("done")
```

```python
### Ignore capitalization and remove punctuation

wordCount = defaultdict(int)
punctuation = set(string.punctuation)
stemmer = PorterStemmer()
for d in data:
    w = stemmer.stem(d.lower())
    if not w in stopWords:
        wordCount[w] += 1
print(len(data))
print(len(wordCount))
print(wordCount)
```

```python
# nomalization
sorted_wordCount = sorted(wordCount.items(), key = lambda kv: kv[1])
for key, value in sorted_wordCount:
    value = value/len(data)
    print ("%s: %s" %(key, value))
```

```python
# plot pdf
import matplotlib.pyplot as plt
k, y = zip(*sorted_wordCount)
x = range(0,len(wordCount))
plt.plot(x, y)
plt.xlabel('number')
plt.ylabel('counts')
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('test2png.png', dpi=100)
fig.set_size_inches(18.5, 10.5, forward=True)
plt.show()
```

```python
#tokenize by sentence
def sentData(fname):
    with open(fname, encoding='utf-8', errors='ignore') as f:
        data = f.read()
    from nltk.tokenize import sent_tokenize
    sentData = sent_tokenize(data)
    return sentData

print("Reading data...")
data = sentData('T1.txt')
print(data)
print("done")
```

```python
# segmentation by line
def lineData(fname):
    data = []
    with open(fname, encoding='utf-8', errors='ignore') as f:
        for line in f:
            data.append(line.lower())
    return data

print("Reading data...")
data = lineData('T1.txt')
print(data)
print("done")
```

```python
def sentLine(line):
    from nltk.tokenize import sent_tokenize
    sentData = sent_tokenize(line)
    return sentData

for line in data:
    if not (line[0] == 'p' or line[0] == 'c'):
        print(line)
    else:
        print(sentLine(line))
        
```
